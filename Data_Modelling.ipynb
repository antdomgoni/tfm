```python
#Definción de variables
df_id = pd.DataFrame()
df_info = pd.DataFrame()
df_tweets = pd.DataFrame()
fact_following = pd.DataFrame()
```

```python
#Lectura en ficheros de los IDS de Twitter
def readID(dataframe,file):
    dataframe = pd.read_csv(file,sep=',')
    return dataframe

following_ids=readID(df_id,'followings_id_hipercor.csv')
following_ids
followers_ids=readID(df_id,'followers_id_hipercor.csv')
followers_ids
```

```python
#Lectura en ficheros de los IDS de Twitter
def readInfo(dataframe,file):
    dataframe = pd.read_csv(file,sep=',')
    return dataframe


df_following_info=readInfo(df_info,'followings_info_hipercor.csv')
df_following_info
df_followers_info=readInfo(df_info,'followers_info_hipercor.csv')
df_followers_info
```

```python
#Lectura en ficheros de los IDS de Twitter
def readTweet(dataframe,file):
    dataframe = pd.read_csv(file,sep=',')
    return dataframe


df_tweets=readTweet(df_tweets,'info_tweets_hipercor.csv')
df_tweets
```

 # Formatting Data

```python
df_following_info.columns=['id_account','name_account','screen_name_account','created_at_account','location_account','followers_account','friends_account','description_account']
df_following_info
```

```python
df_followers_info.columns=['id_account','name_account','screen_name_account','created_at_account','location_account','followers_account','friends_account','description_account']
df_followers_info
```

```python
df_tweets.columns=['created_at_tweet','id_str','id_account','text']
df_tweets
```

# Modelling Data

```python
#Cruce de tablas para construir el tablón final con la información

fact_following=pd.merge(df_following_info, df_tweets,on=['id_account'],how='inner')
fact_following
```

```python
#Cruce de tablas para construir el tablón final con la información

fact_followers=pd.merge(df_followers_info, df_tweets,on=['id_account'],how='inner')
fact_followers
```
