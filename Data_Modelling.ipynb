```python
#Definci√≥n de variables
df_id = pd.DataFrame()
df_info = pd.DataFrame()
df_tweets = pd.DataFrame()
fact_following = pd.DataFrame()
```

```python
#Lectura en ficheros de los IDS de Twitter
def readID(dataframe,file):
    dataframe = pd.read_csv(file,sep=',',usecols=['id_Following'])
    return dataframe

following_ids=readID(df_id,'followings_id.csv')
following_ids
```

```python
#Lectura en ficheros de los IDS de Twitter
def readInfo(dataframe,file):
    dataframe = pd.read_csv(file,sep=',', usecols=['id','name','screen_name','created_at','location','followers_count','friends_count','description'])
    return dataframe


df_following_info=readInfo(df_info,'followings_info.csv')
df_following_info
```

```python
#Lectura en ficheros de los IDS de Twitter
def readTweet(dataframe,file):
    dataframe = pd.read_csv(file,sep=',', usecols=['id_tweet','id_user','text','location','created_at'])
    return dataframe


df_following_info_tweets=readTweet(df_tweets,'followings_info_tweets.csv')
df_following_info_tweets
```

 # Formatting Data

```python
df_following_info.columns=['id_account','name_account','screen_name_account','created_at_account','location_account','followers_account','friends_account','description_account']
df_following_info
```

```python
df_following_info_tweets.columns=['id_tweet','id_account','text_tweet','location_tweet','created_at_tweet']
df_following_info_tweets
```

# Modelling Data

```python
#Cruce de tablas para construir el tabl√≥n final con la informaci√≥n

fact_following=pd.merge(df_following_info, df_following_info_tweets, how='left',
                       left_on=['id_account'], 
                       right_on=['id_account'])
fact_following
```

```python
fact_following.sort_values(by=['id_account','created_at_tweet'])
```
