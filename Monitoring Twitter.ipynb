## Obtención de todos los tweets de un Hashtag de un día (n-1)

```python
data = []
search_words = "#cabify"
date_since = "2020-03-20"

#En caso de eliminar los retweets
#new_search = search_words + " -filter:retweets"

tweets = tweepy.Cursor(api.search,q=search_words,lang="es",since=date_since).items()

for tweet in tweets:
    info_tweet = {}    
    info_tweet["created_at"]=tweet.created_at
    info_tweet["id_str"]=tweet.id_str
    info_tweet["user"]=tweet.user.id
    info_tweet["text"]=tweet.text
    data.append(info_tweet)

df_tweets = pd.DataFrame(data).sort_values("created_at")
df_tweets    
```

## Obtención de todos los tweets de un Hashtag de un rango de días


```python
data = []
search_words = "#cabify"
date_from = "2020-03-01"
date_to = "2020-05-03"

#En caso de eliminar los retweets
#new_search = search_words + " -filter:retweets"

tweets = tweepy.Cursor(api.search,q=search_words,lang="es",since=date_from, until=date_to).items()
for tweet in tweets:
    info_tweet = {}    
    info_tweet["created_at"]=tweet.created_at
    info_tweet["id_str"]=tweet.id_str
    info_tweet["user"]=tweet.user.id
    info_tweet["text"]=tweet.text
    data.append(info_tweet)

df_tweets = pd.DataFrame(data).sort_values("created_at").drop_duplicates()
df_tweets.style.set_properties(subset=['text'], **{'width': '300px'})
df_tweets    
```

## Guardado de tweets en fichero CSV

```python
#Guardado en ficheros de los tweets por usuario de Twitter
def saveTweets(dataframe,file):
    dataframe.to_csv(file, index=False)

saveTweets(df_tweets,'info_tweets_cabify_espana_03052020.csv')
```
